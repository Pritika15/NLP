```python
#preprocessing
def is_token_allowed(token):
  return bool(
      token
       and str(token).strip()
      and not token.is_stop
      and not token.is_punct
  )

def preProcess(token):
  return token.lemma_.strip().lower()

PreprocessText=[
    preProcess(token) for token in text_doc if is_token_allowed(token) and not token.is_punct
]    

PreprocessText
```

```python
#Address
def extract_address(nlp_doc):
  pattern=[
      {"POS":"PROPN"},{"ORTH":"-"},{"POS":"VERB"}
  ]
  matcher.add("Address",[pattern])
  matches=matcher(nlp_doc)
  for _,start,end in matches:
    span=nlp_doc[start:end]
    yield span.text

next(extract_address(text_doc))    
```

```python
print([token.text for token in small_text[5].lefts]) #left of the word
print([token.text for token in small_text[5].rights]) #right of the word
print(small_text[5].nbor(-1)) #prev
print(small_text[5].nbor()) #next
```
